```python
dr1['probsubircorr'] = dr1['probsubircorr'].replace(7777, np.nan)
dr5['probsubircorr'] = dr5['probsubircorr'].replace(7777, np.nan)
dr15['probsubircorr'] = dr15['probsubircorr'].replace(7777, np.nan)
```

```python
dr1['probsubircorr'] = dr1['probsubircorr'].fillna(np.mean(dr1['probsubircorr']))
dr5['probsubircorr'] = dr5['probsubircorr'].fillna(np.mean(dr5['probsubircorr']))
dr15['probsubircorr'] = dr15['probsubircorr'].fillna(np.mean(dr15['probsubircorr']))
```

```python
dr1.describe()
```

```python
# Criar uma função para fazer essa substituição do 777 para os valores médios
def replace7777toMean(df, col_name='probsubircorr'):
    df[col_name] = df[col_name].replace(7777, np.nan)
    df[col_name] = df[col_name].fillna(np.mean(df[col_name]))
    return df
```

```python
dr1 = replace7777toMean(dr1,'probsubirpondcorrpessos')
```

```python
dr1.describe()
```

```python
dr5 = replace7777toMean(dr5,'probsubircorr')
dr15 = replace7777toMean(dr15,'probsubircorr')

dr5 = replace7777toMean(dr5,'probsubirpondcorrpessos')
dr15 = replace7777toMean(dr15,'probsubirpondcorrpessos')
```
