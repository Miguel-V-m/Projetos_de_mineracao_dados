# Lendo Páginas HTML com Python

O Python possui a capacidade de ler páginas HTML disponibilizadas na Web carregando tendo acesso ao código fonte das mesmas. Uma das formas de realizar essa tarefa é por meio da utilização da biblioteca "requests". Neste documento será apresentada um dos inúmeros recursos dessa biblioteca que é a leitura de páginas HTML utilizando-se chamadas do tipo "GET".

Para pode utilizar esta biblioteca é preciso inicialmente importa-la com a instrução a seguir.

```python
import requests
```

Após a importação da biblioteca, chamadas "GET" para a leitura do código fonte de podem ser feitas confome é mostrado a seguir.

```python
html = requests.get("http://labdeeletronica.com.br").text
```

A instrução acima faz uma chamada ao site "http://labdeeletronica.com.br" e o código fonte da página HTML hospedada no mesmo é armazenado na variável "html".

Para confirmar se a variável HTML realmente foi carregada com o código fonte da página solicitada basta imprimir seu conteúdo.

```python
print(html)
```
